# Import modules

```python
import pandas as pd
import re
```

# Process data

```python
# train data
train_df = pd.read_csv('data/train.csv')
train_df
```

```python
# process data
# replace male and female
train_df.replace(['male', 'female'], [1,0], inplace=True)
# remove passenger id
train_df.drop(labels=['PassengerId', 'Name'], axis=1, inplace=True)
# apply hot encoding for Embarked
train_df = pd.get_dummies(train_df, columns=['Embarked'])
# drop Survived
y_train_df = train_df.Survived
x_train_df = train_df.drop(labels=["Survived"], axis=1, inplace=False)
```

```python
x_train_df
```

```python
# extract cabin letters and apply one hot encoding
Cabin_letter = []

for item in x_train_df.Cabin:
    if type(item) == str:
        pattern = '([A-Z])'
        letter = re.findall(pattern, str(item), flags=1)[0]
        Cabin_letter.append(letter)
    else:
        Cabin_letter.append(None)

x_train_df["Cabin_letter"] = Cabin_letter
x_train_df = pd.get_dummies(x_train_df, columns=['Cabin_letter'])
```

```python
# extract ticket letters and apply one hot encoding
Ticket_letter = []

for item in x_train_df.Ticket:
    
    split = item.split(" ")
    if len(split) >= 2:
        Ticket_letter.append(split[0])
    else:
        Ticket_letter.append(None)
        
x_train_df["Ticket_letter"] = Ticket_letter
x_train_df = pd.get_dummies(x_train_df, columns=['Ticket_letter'])
```

```python
x_train_df.drop(labels=['Ticket', 'Cabin'], axis=1, inplace = True)

x_train_df
```
